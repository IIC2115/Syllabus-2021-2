```python
import sqlite3
```

## Consulta 1

```python
connection = sqlite3.connect('data.db')
cursor = connection.cursor()
consulta= """SELECT Profesores.nombre, Cursos.nombre, COUNT(*) AS estudiantes FROM Profesores, Inscritos, Cursos 
            WHERE Cursos.nombre= Inscritos.nombre_curso AND Cursos.id_profesor=Profesores.id
            GROUP BY Cursos.nombre
            ORDER BY estudiantes DESC
            LIMIT 1
            """
cursor.execute(consulta)

print(cursor.fetchall())

connection.commit()
connection.close()
```

## CONSULTA 3

```python
connection = sqlite3.connect('data.db')
cursor = connection.cursor()
consulta= """SELECT DISTINCT Tabla2.id_estud, Tabla2.nombre_cur, Tabla2.horarios_2, Cursos.nombre, Estudiantes.num, Cursos.horario
            FROM (SELECT Estudiantes.num as id_estud , Cursos.horario as horarios_2, Cursos.nombre as nombre_cur FROM Inscritos, Cursos, Estudiantes 
            WHERE Estudiantes.num=Inscritos.num_est AND Cursos.nombre=Inscritos.nombre_curso) as Tabla2, Inscritos, Cursos, Estudiantes 
            WHERE Estudiantes.num=Inscritos.num_est AND Cursos.nombre=Inscritos.nombre_curso 
            AND Tabla2.id_estud= Estudiantes.num AND Cursos.nombre<> Tabla2.nombre_cur
            """
cursor.execute(consulta)

for curso in cursor.fetchall():
    print(curso)

connection.commit()
connection.close()
```

## Consulta 4

```python
connection = sqlite3.connect('data.db')
cursor = connection.cursor()
consulta= """SELECT Estudiantes.major, COUNT(DISTINCT Cursos.sala) as num_salas FROM Inscritos, Cursos, Estudiantes 
            WHERE Estudiantes.num=Inscritos.num_est AND Cursos.nombre=Inscritos.nombre_curso
            GROUP BY Estudiantes.major
            ORDER BY num_salas DESC
            LIMIT 1
            """
cursor.execute(consulta)

for curso in cursor.fetchall():
    print(curso)

connection.commit()
connection.close()
```
